### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
"""Figure: Det är hela den ritbara ytan eller canvasen där dina plotter visas. Tänk på det som ett fönster eller ett papper på vilket du ritar flera plotter. Det kan innehålla flera subplots (axlar).

Axes: Detta är den del av figuren där data plottas. En Figure kan ha flera Axes (subplots). Axes inkluderar både de faktiska plot-ytorna och dekorativa element som axel etiketter, titlar och ticks.

Axis: Dessa är de enskilda axlarna (x och y) inom Axes. Varje Axes-objekt har en x-axel och en y-axel, som styr hur data är skalad och presenterad.

Artist: Detta är alla de grundläggande elementen i en plot, som linjer, text, etiketter, och former. Allting du ser på en plot är en Artist-objekt, och dessa kan läggas till och anpassas för att skapa den visuella representationen av din data.

Enkelt uttryckt, en Figure innehåller en eller flera Axes, var och en med sina Axis-komponenter, och alla visuella element på dessa axlar är Artists."""
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data till scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data till andra grafen med bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# Skafa figure med två subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))

# Scatter plot in det fårsta subplot
ax1.scatter(random_data_x, random_data_y, alpha=0.5)
ax1.plot(x, y, color='red')
ax1.set_title('Scatter Plot')
ax1.set_xlabel('Random X')
ax1.set_ylabel('Random Y')

# Bar plot oin det  andra subplot
ax2.bar(names, values, color=['purple', 'red', 'orange', 'yellow', 'green'])
ax2.set_title('Bar Plot')
ax2.set_xlabel('Fruit')
ax2.set_ylabel('Quantity')

# resultat två graf i en figur,för att undviks överlappar# två graf varandra
#  vi skriver koden som plt.tight_layout() för att automatisk jutserin
plt.tight_layout()  
plt.show()
```
